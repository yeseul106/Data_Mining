```python
# 짜장면 : food_1, 짬뽕 : food_2 csv 파일 불러오기
import pandas as pd
import numpy as np

food_1 =  pd.read_csv('/content/drive/MyDrive/데이터마이닝 수업/food_1.csv', encoding='cp949')
food_2 = pd.read_csv('/content/drive/MyDrive/데이터마이닝 수업/food_2.csv', encoding='cp949')
food_1.head(5)
```

```python
food_2.head(5)
```

```python
# 품목별 항목은 필요 없기 때문에 제거
food_1 = food_1.drop(columns=['품목별'])
food_1.head(1)
```

```python
food_2 = food_2.drop(columns=['품목별'])
food_2.head(1)
```

```python
food_1_np = np.array(food_1)
food_1_np.shape
```

```python
food_2_np = np.array(food_2)
food_2_np.shape
```

```python
food_1_np[0]
```

```python
# 년도별 평균 지수를 나타내기 위해 년도별로 평균을 나타내기
total_year = 11 # 2000-2010년 (11년)
start_index = 1 # 맨 앞에는 '시도별' 열 존재
year_index = 2000

for year in range(total_year):
  sum=0
  for month in range(12):
    sum += food_1_np[0][year*12 + start_index+month] # 일단 전국만
  # print(sum)
  mean = sum / 12
  food_1[str(year_index+year)] = mean
```

```python
food_1.head(1)
```

```python
# 년도별 평균 지수를 나타내기 위해 년도별로 평균을 나타내기
total_year = 11 # 2000-2010년 (11년)
start_index = 1 # 맨 앞에는 '시도별' 열 존재
year_index = 2000

for year in range(total_year):
  sum=0
  for month in range(12):
    sum += food_2_np[0][year*12 + start_index+month] # 일단 전국만
  # print(sum)
  mean = sum / 12
  food_2[str(year_index+year)] = mean
```

```python
food_2.head(1)
```

```python
# 평균 지수 부분만 자르기
food_1_mean = np.array(food_1)
food_1_mean = food_1_mean[0,-11:]
print(food_1_mean)
print(food_1_mean.shape)

food_2_mean = np.array(food_2)
food_2_mean = food_2_mean[0,-11:]
print(food_2_mean)
print(food_2_mean.shape)
```

```python
# 그래프 그리기
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(20,10))
plt.plot(food_1_mean, label='jajang')
plt.plot(food_2_mean, label='jambbong')

plt.legend(['jajang','jambbong'])
```
